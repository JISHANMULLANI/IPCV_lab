```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/Exp 3 hd.jpeg')
```

```python
cv2_imshow(img)
```

```python
total_pixel=img.shape[0]*img.shape[1]
print(total_pixel)
```

```python
#2. Grayscale Conversion
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_bgr = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

cv2_imshow(np.hstack((img, gray_bgr)))
```

```python
#3. Gaussian Blur
blur = cv2.GaussianBlur(img, (15, 15), 0)
cv2_imshow(np.hstack((img, blur)))
```

```python
#4. Edge Detection (Canny)
edges = cv2.Canny(blur, 30, 90)
edges_bgr = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)

cv2_imshow(np.hstack((img, edges_bgr)))
```

```python
#8. Image Rotation
rot = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
rot = cv2.resize(rot, (img.shape[1], img.shape[0]))

cv2_imshow(np.hstack((img, rot)))
```
